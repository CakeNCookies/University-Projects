### 1. Define a python method that takes as input an RGB image and returns a histogram with bin values between 0 ans 255. The method shall return 3 histograms (1 for each channel)

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('lena.png')
#img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# img1 = img[:,:,::-1]

plt.imshow(img[:,:,::-1])
```

```python
def hst(im):
    b,g,r = cv.split(im)
    b = b.flatten()
    g = g.flatten()
    r = r.flatten()
    b_hist = plt.hist(b, bins=range(0,256))
    plt.title('B Channel')
    plt.show()
    g_hist = plt.hist(g, bins=range(0,256))
    plt.title('G Channel')
    plt.show()
    r_hist = plt.hist(r, bins=range(0,256))
    plt.title('R Channel')
    plt.show()
    return b_hist, g_hist, r_hist


hst(img)
```

### 2. Define a python method that can manipulate the intensity of X gray values (chosen by the user) in the input image (channel by channel). The X values will be set to 0 id X < T or 255 if X > T. (T=125). Return the histogram of the original image, the histogram of the processed image, and the final RGB image.

```python
def manip(I):
    val = int(input('please enter an integer: '))
    b,g,r = cv.split(I)
    for i in range(b.shape[1]):
            for j in range(b.shape[0]):
                if b[i,j] == val:
                    if val < 125:
                        b[i,j] = 0
                    else:
                        b[i,j] = 255
    for i in range(g.shape[1]):
            for j in range(g.shape[0]):
                if g[i,j] == val:
                    if val < 125:
                        g[i,j] = 0
                    else:
                        g[i,j] = 255
    for i in range(r.shape[1]):
            for j in range(r.shape[0]):
                if r[i,j] == val:
                    if val < 125:
                        r[i,j] = 0
                    else:
                        r[i,j] = 255

    new_img = cv.merge((b,g,r))
    plt.imshow(new_img[:,:,::-1])
    plt.show()
    
    return hst(I), hst(new_img), new_img
                    
                
manip(img)          
```

### 3. Define a method that can normalize the values of an RGB image between values whithin ranges [a,b] chosen by the user. Process the channels individually. 

```python
def norm(I, s, f):
    b,g,r = cv.split(I)
    maxb = max(b.flatten())
    maxg = max(g.flatten())
    maxr = max(r.flatten())

    minb = min(b.flatten())
    ming = min(g.flatten())
    minr = min(r.flatten())

    for i in range(b.shape[1]):
        for j in range(b.shape[0]):
            if b[i,j] >= s and b[i,j] <= f:
                b[i,j] = 255* ((b[i,j] - minb) / (maxb - minb))

    
    for i in range(g.shape[1]):
        for j in range(g.shape[0]):
            if g[i,j] >= s and g[i,j] <= f:
                g[i,j] = 255* ((g[i,j] - ming) / (maxg - ming))

    for i in range(r.shape[1]):
        for j in range(r.shape[0]):
            if r[i,j] >= s and r[i,j] <= f:
                r[i,j] = 255* ((r[i,j] - minr) / (maxr - minr))

    new_img = cv.merge((b,g,r))
    plt.imshow(new_img[:,:,::-1])
    return new_img
```
